4. High-Quality output in many formats

```python
# comput the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave form')

plt.plot(x,y)
plt.show()
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

### Bar plot

```python
## Bar plot

x = [2,4,8,10] 
y = [11,7,16,9]  

x2 = [4,9,11,6] 
y2 = [6,15,7,4] 
plt.bar(x, y,color='r') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

Above bar graph representing two graph so like this we can also compare the values of more then one graph

## Box Plot using Matplotlib


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

- There are three different array in data set so we have three different boxplot.
- There are five lines in a boxplot :

     lowest line represent 0 percentile
     
     2nd for botton represent 25 percentile
     
     3rd i.e is centeral line for 50 percentile
     
     4th line represent 75 percentile
     
     and last i.e. 5th line represent 100 percentile
     
     
- the circle are basically are outliers (it comes when we randomly select the values)

## Pie chart

```python

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
# size is based on cumulative total value
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.1, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
